```python
#Setup and loading data
import pandas as pd

# Load the dataset
# If the file is in your main Colab folder, the path is 'bank_loan.csv'
try:
    df = pd.read_csv('/bank_loan.csv')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("File not found. Please upload 'bank_loan.csv' to the Colab sidebar.")
```

```python
#Data Inspection
# View the first 5 rows to see the data structure
print("--- First 5 Rows ---")
display(df.head())

# Check for data types and non-null counts
print("\n--- Data Structure and Info ---")
df.info()

# Check for the number of missing values in each column
print("\n--- Missing Values Count ---")
print(df.isnull().sum())
```

```python
#Data Cleaning
# 1. Handling Missing Values
# Strategy: If the missing data is in a critical column (like 'Loan_ID'), we drop it.
# For numerical columns like 'LoanAmount', we fill with the median to avoid outliers.
if 'LoanAmount' in df.columns:
    df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())

# 2. Removing Duplicates
# Strategy: Duplicate entries can skew statistical analysis and lead to double-counting.
initial_rows = len(df)
df.drop_duplicates(inplace=True)
print(f"Removed {initial_rows - len(df)} duplicate rows.")

# 3. Fixing Inconsistencies
# Strategy: Ensure categorical data is uniform (e.g., changing 'YES'/'yes' to 'Yes')
# Here we strip whitespace from any string columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

print("\nCleaning complete.")
```

```python
#Save cleaned dataset
# Save to a new CSV file without the index column
df.to_csv('bank_loan_cleaned.csv', index=False)

print("Cleaned dataset saved as 'bank_loan_cleaned.csv'.")
```
